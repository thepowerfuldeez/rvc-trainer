```python
import pandas as pd
df = pd.read_csv("/home/george/datasets/singing_voice/acappella/full_dataset.csv")
```

```python
from pytube import YouTube
from pathlib import Path
from tqdm.auto import tqdm
import subprocess
```

```python
out_dir = Path("/home/george/datasets/singing_voice/acappella/youtube_audios/")
out_wav_dir = Path("/home/george/datasets/singing_voice/acappella/cut_audios/")
out_wav_dir.mkdir(exist_ok=True, parents=True)
out_dir.mkdir(exist_ok=True, parents=True)
```

```python
speakers_counts = dict(df.loc[df['Length (s)'] > 60].value_counts('Singer'))
```

```python
speakers_dict = {}
k = 0
for n, c in speakers_counts.items():
    if c > 1:
        speakers_dict[n] = k
        k += 1
```

```python
total_length = 0

for _, row in tqdm(list(df.iterrows())):
    if row['Length (s)'] > 60:
        link = row['Link']
        try:
            yt = YouTube(link)
            stream = yt.streams.filter(only_audio=True).order_by('abr').desc().first()
            suffix = stream.mime_type.split("/")[1]

            speaker = row['Singer']
            if speaker in speakers_dict:
                spk_id = speakers_dict[speaker]
            else:
                continue
            
            filename = f"{spk_id}_{row['Song Name']}.{suffix}".replace(" ", "_").replace("(", "").replace(")", "")
            stream.download(output_path=out_dir, filename=filename)

            out_path = out_dir / filename
    
            start = str(row['Init']).replace(".", ":")
            end = str(row['Fin']).replace(".", ":")
            new_path = str(out_wav_dir / out_path.with_suffix('.wav').name)
    
            subprocess.run(f"ffmpeg -i {str(out_path)} -ss {start} -to {end} {new_path}", 
                           check=True, shell=True)
            total_length += row['Length (s)']
        except Exception as e:
            print(e)
            print(row['Song Name'])
            continue
```

```python
items = soup.find_all("div", class_="c-mtk__artist m-container m-container--artist")

for j, artist_it in tqdm(enumerate(items[276:], 276)):
    
    try:
        for line_i, line_it in enumerate(artist_it.find_all("li", class_="m-mtk-track")):
            out_path = out_dir / f"{j}" / f"{line_i}.zip"
            out_path.parent.mkdir(exist_ok=True, parents=True)
            
            it = line_it.find_all("li", class_="m-mtk-download")[-2]
            it2 = it.find("span", class_="m-mtk-download__links")
            if it2:
                link = it2.find('a')['href']
                out_path.write_bytes(requests.get(link).content)
    except Exception as e:
        print(e)
```
